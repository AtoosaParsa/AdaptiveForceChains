```python
import pickle
import matplotlib.pyplot as plt
from simulator import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

import seaborn as sns

matplotlib.rcParams['axes.linewidth'] = 1.5
sns.set_style("ticks", {"ytick.major.size": 5, "xtick.major.size": 5, 'font.family': 'serif', 'font.serif': 'Times New Roman'})

runs = c.RUNS
gens = c.numGenerations
fitnesses = np.zeros([runs, gens])
temp = []
individuals = []
with open('/cluster/tufts/levinlab/aparsa01/data/forceChains/Vshape/savedRobotsAfpoSeed.dat', "rb") as f:
    for r in range(1, runs+1):
        for g in range(1, gens+1):
            try:
                if g == gens:
                    individuals.append(pickle.load(f))
                    temp.append(individuals[0].fitness)
                    #print(individuals[0].genome)
                else:
                    temp.append(pickle.load(f).fitness)
            except EOFError:
                break
        fitnesses[r-1] = temp
        temp = []
f.close()

mean_f = np.mean(fitnesses, axis=0)
std_f = np.std(fitnesses, axis=0)

bestSolution = mean_f[-1]
print(f"best solution: {bestSolution}")

fig, ax = plt.subplots(figsize=(7, 5))

ax.plot(list(range(1, gens+1)), mean_f, color='tomato', linewidth=2, label='Best EA Solution', linestyle='solid')
ax.fill_between(list(range(1, gens+1)), mean_f-std_f, mean_f+std_f, color='tomato', alpha=0.2, linewidth=1)

temp = []
fitnesses = np.zeros([runs, gens])
with open('/cluster/tufts/levinlab/aparsa01/data/forceChains/Vshape/avgFitnessAfpoSeed.dat', "rb") as f:
    for r in range(1, runs+1):
        for g in range(1, gens+1):
            try:
                temp.append(pickle.load(f))
            except EOFError:
                break
        fitnesses[r-1] = temp
        temp = []
f.close()

mean_f = np.mean(fitnesses, axis=0)
std_f = np.std(fitnesses, axis=0)

avgSolution = -mean_f[-1]

ax.plot(list(range(1, gens+1)), -mean_f, color='royalblue', linewidth=2, label='Average EA Solution', linestyle=(0, (5, 1)))
ax.fill_between(list(range(1, gens+1)), -(mean_f-std_f), -(mean_f+std_f), color='royalblue', alpha=0.2, linewidth=1)
#ax.grid(which='minor', color='gray', linestyle=':', linewidth=0.2)
#ax.grid(which='major', color='gray', linestyle='-', linewidth=0.3)
ax.set_axisbelow(True)
ax.minorticks_on()
legend = ax.legend(loc='lower right', fontsize=14)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
ax.set_ylim([0.13, 0.49])
ax.set_xlabel('Generations', fontsize=16) #, fontweight='semibold')
ax.set_ylabel(r'$\mathcal{O}_0 = F_{o,3} \: / \: F_i$', fontsize=16) #, fontweight='semibold')
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)

plt.savefig('optimization.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.savefig('optimization.pdf', bbox_inches='tight', format='pdf', dpi=600, transparent=True)

plt.show()
plt.close(fig)
```
