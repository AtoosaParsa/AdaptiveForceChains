```python
import pickle
import matplotlib.pyplot as plt
from simulator import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

import seaborn as sns

matplotlib.rcParams['axes.linewidth'] = 1.5
sns.set_style("ticks", {"ytick.major.size": 5, "xtick.major.size": 5, 'font.family': 'serif', 'font.serif': 'Times New Roman'})

runs = [1, 2, 3]
gens = c.numGenerations
pops = c.popSize

best_fitnesses = np.zeros([len(runs), gens])
for r in np.arange(len(runs)):
    with open(f'bests{runs[r]}.dat', "rb") as f:
        for g in range(1, gens+1):
            best_fitnesses[r-1, g-1] = pickle.load(f).fitnesses[0] ** (1. / 2)
    f.close()

mean_f = np.mean(best_fitnesses, axis=0)
std_f = np.std(best_fitnesses, axis=0)

bestSolution = mean_f[-1]
print(bestSolution)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(list(range(1, gens+1)), mean_f, color='tomato', linewidth=2, label='Best Solution', linestyle='solid')
ax.fill_between(list(range(1, gens+1)), mean_f-std_f, mean_f+std_f, color='tomato', alpha=0.2, linewidth=1)

avg_fitnesses = np.zeros([len(runs), gens-1])
for r in np.arange(len(runs)):
    for g in range(1, gens):
        with open(f'gens{runs[r]}/gen{g}.dat', "rb") as f:
            genomes = pickle.load(f)
            f_sum = 0
            for ind in genomes:
                f_sum = f_sum + genomes[ind].fitnesses[0] ** (1. / 2)
        f.close()
        avg_fitnesses[r-1, g-1] = f_sum / pops
#for r in np.arange(len(runs)):
#    with open(f'avgFitness{runs[r]}.dat', "rb") as f:
#        for g in range(1, gens+1):
#            avg_fitnesses[r-1, g-1] = pickle.load(f)[0]
#    f.close()

mean_f = np.mean(avg_fitnesses, axis=0)
std_f = np.std(avg_fitnesses, axis=0)

avgSolution = mean_f[-1]

ax.plot(list(range(1, gens)), mean_f, color='royalblue', linewidth=2, label='Average Solution', linestyle=(0, (5, 1)))
ax.fill_between(list(range(1, gens)), (mean_f-std_f), (mean_f+std_f), color='royalblue', alpha=0.2, linewidth=1)
#ax.grid(which='minor', color='gray', linestyle=':', linewidth=0.2)
#ax.grid(which='major', color='gray', linestyle='-', linewidth=0.3)
ax.set_axisbelow(True)
ax.minorticks_on()
legend = ax.legend(loc='lower right', fontsize=14)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
ax.set_ylim([0.00165, 0.0033])
ax.set_xlabel('Generations', fontsize=16) #, fontweight='semibold')
ax.set_ylabel(r'$\mathcal{O}_2 = \sqrt{F_{o,2} \; F_{o,4}} \: / \: F_i$', fontsize=16) #, fontweight='semibold')
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)

plt.savefig('optimization.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```

```python
import pickle
import matplotlib.pyplot as plt
from simulator6 import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

font = {'family' : 'sans-serif'}#,'weight' : 'semibold'}  
plt.rc('font', **font) 
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

#plt.rcParams["font.family"] = "sans-serif"
#plt.rcParams['figure.dpi'] = 600
#plt.rcParams['axes.linewidth'] = 1.5
#plt.rcParams['figure.figsize'] = (15.0, 10.0)
#plt.rcParams['savefig.bbox'] = 'tight'

runs = c.RUNS
gens = c.numGenerations
fitnesses = numpy.zeros([runs, gens])
temp = []
individuals = []
with open('savedRobotsAfpoSeed.dat', "rb") as f:
    for r in range(1, runs+1):
        for g in range(1, gens+1):
            try:
                if g == gens:
                    individuals.append(pickle.load(f))
                    temp.append(individuals[0].fitness)
                else:
                    temp.append(pickle.load(f).fitness)
            except EOFError:
                break
        fitnesses[r-1] = temp
        temp = []
f.close()

mean_f = numpy.mean(fitnesses, axis=0)
std_f = numpy.std(fitnesses, axis=0)

# set up the axis and subplots
figsize = 10
fig = plt.figure(figsize=(1*figsize, 0.35*figsize))
subfigs = fig.subfigures(nrows=1, ncols=2, wspace=0.05)
#subfig.suptitle(f'sample {num+1}: {initial_positions}, RMSE={rmse}', fontsize=14, y=0.95, fontweight='normal', color='black')
#subfig.subplots_adjust(top=0.8, wspace=0, hspace=0.1)
main_ax = subfigs[0].subplots(nrows=1, ncols=1) #, sharey=True, sharex=True)

main_ax.plot(list(range(1, gens+1)), mean_f, color='red', linewidth=1.5, label='Best')
main_ax.fill_between(list(range(1, gens+1)), mean_f-std_f, mean_f+std_f, color='red', alpha=0.3, linewidth=1)

## plot the average fitness
temp = []
fitnesses = numpy.zeros([runs, gens])
with open('avgFitnessAfpoSeed.dat', "rb") as f:
    for r in range(1, runs+1):
        for g in range(1, gens+1):
            try:
                temp.append(pickle.load(f))
            except EOFError:
                break
        fitnesses[r-1] = temp
        temp = []
f.close()

mean_f = numpy.mean(fitnesses, axis=0)
std_f = numpy.std(fitnesses, axis=0)

main_ax.plot(list(range(1, gens+1)), -mean_f, color='blue', linewidth=1.5, label='Average')
main_ax.fill_between(list(range(1, gens+1)), -(mean_f-std_f), -(mean_f+std_f), color='cornflowerblue', alpha=0.3, linewidth=1)
main_ax.set_axisbelow(True)
main_ax.grid(which='minor', color='gray', linestyle=':', linewidth=0.2)
main_ax.grid(which='major', color='gray', linestyle='-', linewidth=0.3)
main_ax.set_axisbelow(True)
main_ax.minorticks_on()
main_ax.legend(loc='lower right', fontsize=14)
main_ax.set_ylim([0.15, 0.41])
main_ax.set_xlabel('$Generations$', fontsize=14)#, fontweight='bold')
main_ax.set_ylabel(r'$O = \frac{F_o}{F_i}$', fontsize=14)#, fontweight='bold')
main_ax.tick_params(axis='x', which='major', labelsize=12, width=1.5)
main_ax.tick_params(axis='x', which='minor', labelsize=12, width=1)
main_ax.tick_params(axis='y', which='major', labelsize=12, width=1.5)
main_ax.tick_params(axis='y', which='minor', labelsize=12, width=1)

y_hist = subfigs[1].subplots(nrows=1, ncols=1, sharey=True, sharex=True)

# plot the histogram
with open('RS.dat', "rb") as f:
    rs = pickle.load(f)
f.close()
#_, bins, _ = y_hist.hist(x=rs, bins='auto', color='dimgray', alpha=0.8, rwidth=0.85, cumulative=False, orientation='vertical', histtype='stepfilled', density=False)#, label='random') #, grid=True)
rs = np.array(rs)
rs_grid = np.linspace(min(rs), max(rs), 1000)
kde = gaussian_kde(rs, bw_method=0.005 / rs.std(ddof=1))
kdepdf = kde.evaluate(rs_grid)
y_hist.plot(rs_grid, kdepdf, color='red', alpha=0.5, lw=1.5, label='kde')
#y_hist.axvline(x=24.5, ymin=0.0, ymax=1.0, color='k', linewidth=1)
y_hist.minorticks_on()
y_hist.set_xlabel(r'$O = \frac{F_o}{F_i}$', fontsize=14)#, fontweight='bold')
y_hist.set_ylabel('Counts', fontsize=14)
#formatter = matplotlib.ticker.FormatStrFormatter('%1.0f')
#y_hist.yaxis.set_major_formatter(formatter)
#y_hist.tick_params(axis='y', which='major', labelsize=12, width=1.5)
#y_hist.tick_params(axis='y', which='minor', labelsize=12, width=1)
#y_hist.tick_params(axis='x', which='major', labelsize=12, width=1.5)
#y_hist.tick_params(axis='x', which='minor', labelsize=12, width=1)
#y_hist.tick_params(axis='y', which='minor', labelsize=14)
y_hist.legend(loc='upper left', fontsize=14)
y_hist.set_xlim([0.15, 0.41])

# fitting a normal distribution
#mu, std = norm.fit(rs)
#xmin, xmax = y_hist.get_ylim()
#x = numpy.linspace(xmin, xmax, 1000)
#p = norm.pdf(x, mu, std)

#plt.plot(x, p, linewidth=2)
#myText = "Mean={:.2f}, STD={:.2f}".format(mu, std)
#y_hist.hlines(y=0.12, xmin=0, xmax=10, linewidth=1, linestyle='--', color='limegreen', alpha=0.9)
#y_hist.text(0.5, 0.9, myText, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=28, color='limegreen', fontweight='bold')

#fig.tight_layout()
savename = 'V_shape'+'.png'
savepath = savename #os.path.join(savedir, savename)
plt.savefig(savepath, bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```
