```python
import pickle
import matplotlib.pyplot as plt
from simulator import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

import seaborn as sns

matplotlib.rcParams['axes.linewidth'] = 1.5
sns.set_style("ticks", {"ytick.major.size": 5, "xtick.major.size": 5, 'font.family': 'serif', 'font.serif': 'Times New Roman'})

runs = [1, 2, 3]
gens = c.numGenerations
pops = c.popSize

best_fitnesses = np.zeros([len(runs), gens])
for r in np.arange(len(runs)):
    with open(f'data/bests{runs[r]}.dat', "rb") as f:
        for g in range(1, gens+1):
            best_fitnesses[r-1, g-1] = pickle.load(f).fitnesses[0]
    f.close()

mean_f = np.mean(best_fitnesses, axis=0)
std_f = np.std(best_fitnesses, axis=0)

bestSolution = mean_f[-1]
print(bestSolution)

fig, ax = plt.subplots(figsize=(7, 5))

ax.plot(list(range(1, gens+1)), mean_f, color='tomato', linewidth=2, label='Best EA Solution', linestyle='solid')
ax.fill_between(list(range(1, gens+1)), mean_f-std_f, mean_f+std_f, color='tomato', alpha=0.2, linewidth=1)

avg_fitnesses = np.zeros([len(runs), gens-1])
for r in np.arange(len(runs)):
    for g in range(1, gens):
        with open(f'data/gens{runs[r]}/gen{g}.dat', "rb") as f:
            genomes = pickle.load(f)
            f_sum = 0
            for ind in genomes:
                f_sum = f_sum + genomes[ind].fitnesses[0]
        f.close()
        avg_fitnesses[r-1, g-1] = f_sum / pops

mean_f = np.mean(avg_fitnesses, axis=0)
std_f = np.std(avg_fitnesses, axis=0)

avgSolution = mean_f[-1]

ax.plot(list(range(1, gens)), mean_f, color='royalblue', linewidth=2, label='Average EA Solution', linestyle=(0, (5, 1)))
ax.fill_between(list(range(1, gens)), (mean_f-std_f), (mean_f+std_f), color='royalblue', alpha=0.2, linewidth=1)
ax.set_axisbelow(True)
ax.minorticks_on()
legend = ax.legend(loc='lower right', fontsize=14)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
ax.set_ylim([0.10, 0.24])
ax.set_xlabel('Generations', fontsize=16) #, fontweight='semibold')
ax.set_ylabel(r'$\mathcal{O}_1 = \sqrt[3]{F_{o,1} \; F_{o,3} \; F_{o,5}} \: / \: F_i$', fontsize=16) #, fontweight='semibold')
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)

plt.savefig('figures/optimization_O1.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```

```python
from matplotlib.ticker import PercentFormatter

fig, ax = plt.subplots(figsize=(7, 5))

with open('RS/outs.dat', "rb") as f:
    rs = np.array(pickle.load(f))
f.close()

print(np.min(rs))
print(np.max(rs))

# Probability:
_, bins, _ = ax.hist(x=rs, bins=30, weights = (1/len(rs)) * np.ones(len(rs)), color='dimgray', alpha=0.8, rwidth=0.9, cumulative=False, orientation='vertical', density=False, edgecolor='black', linewidth=0.5, label='Gen 0 Distribution')#, grid=True, , histtype='stepfilled')

ax.axvline(x=bestSolution, ymin=0.0, ymax=1, color='tomato', linewidth=2, linestyle='solid', label='Best EA Solution')
ax.axvline(x=avgSolution, ymin=0.0, ymax=1, color='royalblue', linewidth=2, linestyle=(0, (5, 1)), label='Average EA Solution')

ax.minorticks_on()
ax.set_axisbelow(True)
ax.set_xlabel(r'$\mathcal{O}_1 = \sqrt[3]{F_{o,1} \; F_{o,3} \; F_{o,5}} \: / \: F_i$', fontsize=16)
ax.set_ylabel('Probability', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)
legend = ax.legend(loc='upper right', fontsize=13, ncols=1)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
ax.set_xlim([0.10, 0.24])
ax.set_ylim([0, 0.1])

plt.savefig('figures/histogram_O1.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```
