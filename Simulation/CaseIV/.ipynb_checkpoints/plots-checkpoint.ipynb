```python
import pickle
import matplotlib.pyplot as plt
from simulator import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

import seaborn as sns

matplotlib.rcParams['axes.linewidth'] = 1.5
sns.set_style("ticks", {"ytick.major.size": 5, "xtick.major.size": 5, 'font.family': 'serif', 'font.serif': 'Times New Roman'})

runs = [1, 2, 3]
gens = c.numGenerations
pops = c.popSize

best_fitnesses = np.zeros([len(runs), gens-1])
for r in np.arange(len(runs)):
    f_best = 0
    for g in range(1, gens):
        with open(f'data/gens{runs[r]}/gen{g}.dat', "rb") as f:
            genomes = pickle.load(f)
            for ind in genomes:
                temp = genomes[ind].fitnesses[0]
                if temp > f_best:
                    f_best = temp
        f.close()
        best_fitnesses[r-1, g-1] = f_best

mean_f = np.mean(best_fitnesses, axis=0)
std_f = np.std(best_fitnesses, axis=0)

bestSolution = mean_f[-1]

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(list(range(1, gens)), mean_f, color='tomato', linewidth=2, label='Best',  linestyle='solid')
ax.fill_between(list(range(1, gens)), mean_f-std_f, mean_f+std_f, color='tomato', alpha=0.2, linewidth=1)

avg_fitnesses = np.zeros([len(runs), gens-1])
for r in np.arange(len(runs)):
    for g in range(1, gens):
        with open(f'data/gens{runs[r]}/gen{g}.dat', "rb") as f:
            genomes = pickle.load(f)
            f_sum = 0
            for ind in genomes:
                f_sum = f_sum + genomes[ind].fitnesses[0]
        f.close()
        avg_fitnesses[r-1, g-1] = f_sum / pops

mean_f = np.mean(avg_fitnesses, axis=0)
std_f = np.std(avg_fitnesses, axis=0)

avgSolution = mean_f[-1]

ax.plot(list(range(1, gens)), mean_f, color='royalblue', linewidth=2, label='Average', linestyle=(0, (5, 1)))
ax.fill_between(list(range(1, gens)), (mean_f-std_f), (mean_f+std_f), color='royalblue', alpha=0.2, linewidth=1)
ax.set_axisbelow(True)
ax.minorticks_on()
legend = ax.legend(loc='lower right', fontsize=14, ncols=2)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
#ax.set_ylim([0.12, 0.22])
ax.set_xlabel('Generations', fontsize=16) #, fontweight='semibold')
ax.set_ylabel(r'$\mathcal{O}_1 = \sqrt[3]{F_{o,1} \; F_{o,3} \; F_{o,5}} \: / \: F_i$', fontsize=16) #, fontweight='semibold')
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)

plt.savefig('figures/optimization_O1.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```

```python
import topsispy as tp
import pickle
import matplotlib.pyplot as plt
from simulator import simulator
import constants as c
import numpy as np
import matplotlib
from scipy.stats import norm
from scipy.stats import gaussian_kde

import seaborn as sns

matplotlib.rcParams['axes.linewidth'] = 1.5
sns.set_style("ticks", {"ytick.major.size": 5, "xtick.major.size": 5, 'font.family': 'serif', 'font.serif': 'Times New Roman'})

runs = [1, 2, 3]
gens = c.numGenerations
pops = c.popSize

def paretoFront(population):
    pareto_front = []

    for i in population:
        i_is_dominated = False
        for j in population:
            if i != j:
                if population[j].dominates2(population[i]):
                    i_is_dominated = True
                    break
        if not i_is_dominated:
            pareto_front.append(population[i])

    return pareto_front

fig, ax = plt.subplots(figsize=(10, 5))

r = 1

with open(f'data/gens{r}/gen{gens-1}.dat', "rb") as f:
    population = pickle.load(f)
    pf = paretoFront(population)
    print("pf size: "+str(len(pf)))

    for i in population:
        if i == 0:
            ax.scatter(x=population[i].fitnesses[1], y=population[i].fitnesses[0], color=(0, 0, 1, 0), s=40, edgecolors='darkgoldenrod', linewidth=1.5, label='Last Generation')
        else:
            ax.scatter(x=population[i].fitnesses[1], y=population[i].fitnesses[0], color=(0, 0, 1, 0), s=40, edgecolors='darkgoldenrod', linewidth=1.5)
            
    data = []
    for i in pf:
        data.append([i.fitnesses[0], i.fitnesses[1]])

    weights = [0.5, 0.5]
    sign = [1, -1]
    bestIndex_05_05, ranks = tp.topsis(data, weights, sign)

    weights = [0, 1]
    sign = [1, -1]
    bestIndex_0_1, ranks = tp.topsis(data, weights, sign)

    weights = [1, 0]
    sign = [1, -1]
    bestIndex_1_0, ranks = tp.topsis(data, weights, sign)
    
    count = 0
    for i in np.arange(len(pf)):
        if i == bestIndex_05_05:
            ax.scatter(x=pf[i].fitnesses[1], y=pf[i].fitnesses[0], marker='*', color=(0, 0, 0, 0), s=180, edgecolors='magenta', linewidth=3, label='Selected Solution W=[0.5, 0.5]')
            
        elif i == bestIndex_0_1:
            ax.scatter(x=pf[i].fitnesses[1], y=pf[i].fitnesses[0], marker='D', color=(0, 0, 0, 0), s=100, edgecolors='darkorange', linewidth=3, label='Selected Solution W=[1.0, 0.0]')
            
        elif i == bestIndex_1_0:
            ax.scatter(x=pf[i].fitnesses[1], y=pf[i].fitnesses[0], marker='^', color=(0, 0, 0, 0), s=130, edgecolors='blueviolet', linewidth=3, label='Selected Solution W=[0.0, 1.0]')

        else:
            if count == 0:
                    ax.scatter(x=pf[i].fitnesses[1], y=pf[i].fitnesses[0], marker = 's', color=(0, 0, 0, 0), s=70, edgecolors='black', linewidth=2, label='Pareto Front')
            else:
                ax.scatter(x=pf[i].fitnesses[1], y=pf[i].fitnesses[0], marker='s', color=(0, 0, 0, 0), s=70, edgecolors='black', linewidth=2)
            count = count + 1        
f.close()

ax.set_axisbelow(True)
ax.minorticks_on()
handles, labels = plt.gca().get_legend_handles_labels()
#print(labels)
order = [0,1,2,4,3]
legend = ax.legend([handles[idx] for idx in order],[labels[idx] for idx in order], loc='lower right', fontsize=12)
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 1))
#ax.set_ylim([0.12, 0.22])
ax.set_xlabel(r'$\mathcal{O}_4 = N_{C_2^* \rightarrow C_1}$', fontsize=16) #, fontweight='semibold')
ax.set_ylabel(r'$\mathcal{O}_1 = \sqrt[3]{F_{o,1} \; F_{o,3} \; F_{o,5}} \: / \: F_i$', fontsize=16) #, fontweight='semibold')
ax.tick_params(axis='both', which='major', labelsize=14, width=2.5)
ax.tick_params(axis='both', which='minor', labelsize=14, width=1.5)

#plt.title(f"TOPSIS Weights = {weights}", fontsize=16, fontweight='semibold')

plt.savefig('figures/paretoFront.png', bbox_inches='tight', format='png', dpi=600, transparent=True)
plt.show()
plt.close(fig)
```
